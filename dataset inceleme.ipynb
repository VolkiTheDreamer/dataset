```python
import glob
dflist=[]
i=0

for filepath in glob.glob(os.curdir + '/**/*.csv', recursive=True):
    try:
        print(filepath)
        df=pd.read_csv(filepath,encoding = "ISO-8859-1")    
        dflist.append(df)
        i+=1
        print(i)
        if len(df.columns)==1:        
            df=pd.read_csv(filepath,encoding = "ISO-8859-1",sep=";")
        print(len(df))
        df.head()
    except:
        pass
```
